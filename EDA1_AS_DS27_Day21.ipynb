```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('Telco_customer_churn.csv')

# 1. Handle Missing Values

# Check for missing values
print(df.isnull().sum())

# Replace missing values in 'TotalCharges' with the mean
df['TotalCharges'] = df['TotalCharges'].replace(' ', np.nan).astype(float)
df['TotalCharges'].fillna(df['TotalCharges'].mean(), inplace=True)

# 2. Handle Duplicate Data

# Check for duplicate rows
print(df.duplicated().sum())

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Print the cleaned dataset
print(df.head())

# Save the cleaned dataset
df.to_csv('cleaned_Telco_customer_churn.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
df = pd.read_csv('cleaned_Telco_customer_churn.csv')

# Pertanyaan Eksploratif:

# 1. Apakah pelanggan dengan layanan internet yang lebih cepat cenderung lebih loyal (tidak churn)?
# 2. Apakah ada perbedaan signifikan dalam total biaya yang dibayarkan oleh pelanggan yang churn dan yang tidak churn, berdasarkan jenis layanan internet?

# Analisis Mendalam:

# 1. Layanan Internet dan Loyalitas Pelanggan

# Group by InternetService and calculate churn rate
churn_rate_by_internet = df.groupby('InternetService')['Churn'].apply(lambda x: (x=='Yes').mean())
print(churn_rate_by_internet)

# Visualisasi churn rate berdasarkan layanan internet
plt.figure(figsize=(8, 5))
sns.barplot(x=churn_rate_by_internet.index, y=churn_rate_by_internet.values)
plt.xlabel('Internet Service')
plt.ylabel('Churn Rate')
plt.title('Churn Rate by Internet Service')
plt.show()

# Interpretasi:
# - Pelanggan dengan layanan internet 'Fiber optic' memiliki churn rate yang lebih tinggi dibandingkan dengan layanan internet lainnya.
# - Pelanggan dengan layanan internet 'DSL' memiliki churn rate yang lebih rendah dibandingkan dengan layanan internet 'Fiber optic'.
# - Pelanggan dengan layanan internet 'No' memiliki churn rate yang terendah.

# 2. Total Biaya dan Churn Berdasarkan Jenis Layanan Internet

# Calculate average total charges by internet service and churn
avg_total_charges = df.groupby(['InternetService', 'Churn'])['TotalCharges'].mean()

# Group by InternetService and calculate the mean churn rate
churn_rate_by_internet = df.groupby('InternetService')['Churn'].apply(lambda x: (x=='Yes').mean()) 
print(churn_rate_by_internet)

# Visualisasi total biaya rata-rata berdasarkan layanan internet dan churn
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_total_charges.index.get_level_values(0),
            y=avg_total_charges.values,
            hue=avg_total_charges.index.get_level_values(1))
plt.xlabel('Internet Service')
plt.ylabel('Average Total Charges')
plt.title('Average Total Charges by Internet Service and Churn')
plt.show()

# Interpretasi:
# - Pelanggan yang churn dengan layanan internet 'Fiber optic' memiliki total biaya rata-rata yang lebih tinggi dibandingkan dengan pelanggan yang tidak churn.
# - Pelanggan yang churn dengan layanan internet 'DSL' memiliki total biaya rata-rata yang lebih rendah dibandingkan dengan pelanggan yang tidak churn.
# - Pelanggan yang churn dengan layanan internet 'No' memiliki total biaya rata-rata yang paling rendah.

# Kesimpulan:
# - Pelanggan dengan layanan internet 'Fiber optic' cenderung memiliki churn rate yang lebih tinggi dan total biaya rata-rata yang lebih tinggi.
# - Layanan internet 'DSL' memiliki churn rate yang lebih rendah dan total biaya rata-rata yang lebih rendah.
# - Layanan internet 'No' memiliki churn rate yang terendah dan total biaya rata-rata yang paling rendah.
```
