```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

city = ['Malang', 'Pontianak 1', 'Depok', 'Bogor', 'Pontianak 2']
counts = [539, 498, 509, 486, 498 ]
# Adjust the number of labels to match the number of bars (cities)
bar_labels = ['red', 'blue', 'green', 'orange', 'purple']
# Use actual color values instead of colormap names
bar_colors = ['red', 'blue', 'green', 'orange', 'purple']

# Use 'city' as the city names for the x-axis
ax.bar(city, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('Populasi')
ax.set_title('City')

plt.show()
```

# Insight dari Barplot di atas sesuai dengan urutan dataset
1. Populasi tertinggi dan terendah:
*  Kota Malang memiliki populasi tertinggi yaitu 539.
*  Kota Bogor memiliki populasi terendah yaitu 486.

2. Populasi:
* Kota Pontianak muncul dua kali dalam daftar dengan populasi yang sama yaitu 498 pada keduanya. Ini bisa menunjukkan adanya duplikasi data atau pengulangan untuk tujuan tertentu.

3. Distribusi Populasi:
* Populasi di kota-kota ini bervariasi mulai dari 486 hingga 539. Rentang variasi populasi ini memberi gambaran bahwa populasi tersebar di antara kota tersebut.

4. Perbedaan warna:
* Malang : Merah
* Pontianak : Biru
* Depok : Hijau
* Bogor : Oranye
* Pontianak : Ungu

# Penjelasan Kode

1. Data:
* City berisi nama-nama kota
* counts berisi populasi masing-masing kota.
* bar_labels dan bar_colors berisi informasi pada label dan warna di setiap bar.

2. Plotting:
* Fungsi ax.bar() digunakan untuk membuat barplot dengan kota sebagai sumbu x dan populasi sebagai sumbu y.
* Judul sumbu y diatur menjadi 'Populasi' dan judul plot diatur menjadi 'City'

```python
import pandas as pd

# Membaca dataset dari file CSV
df = pd.read_csv('dataset_day6.csv')

# Menampilkan beberapa baris pertama dari DataFrame untuk eksplorasi
print("Beberapa baris pertama dari DataFrame:")
print(df.head(5))
print()
print('==============================================')

# Looping pada kolom
print("Looping pada kolom:")
for col in df.columns:  # Mengiterasi nama-nama kolom
    print(f"Nama Kolom: {col}")
    print(f"Tipe Data: {df[col].dtype}")
    print(f"Beberapa Nilai: {df[col].head(5).tolist()}")
    print()
print('==============================================')

# Looping pada baris untuk 5 data pertama
print("Looping pada baris (5 data pertama):")
for index, row in df.head(5).iterrows():
    print(f"Index: {index}")
    row_dict = row.to_dict()
    print("Data Baris (sebagai dictionary):")
    for key, value in row_dict.items():
        print(f"{key}: {value}")
    print()
```

### While loop

```python
import pandas as pd

# Membaca dataset dari file CSV
df = pd.read_csv('dataset_day6.csv')

# Menampilkan beberapa baris pertama dari DataFrame untuk eksplorasi
print("Beberapa baris pertama dari DataFrame:")
print(df.head(5))
print()

# Inisialisasi nilai awal
i = 0  # Mulai dari baris pertama

# Looping pada baris DataFrame
while i < len(df):  # Akan terus looping jika i kurang dari jumlah baris di DataFrame
    print(f'Urutan output ke : {i+1}')
    print(df.iloc[i])  # Menampilkan baris ke-i dari DataFrame

    if i == 4:  # Kondisi untuk menghentikan loop saat i = 3 (baris ke-4)
        break

    i += 1  # Increment nilai i

print()
```

```python
import pandas as pd

# Membaca dataset dari file CSV
df = pd.read_csv('dataset_day6.csv')

# Menampilkan beberapa baris pertama dari DataFrame untuk eksplorasi
print("Beberapa baris pertama dari DataFrame:")
print(df.head(5))
print()
print('==============================================')


# Looping pada kolom menggunakan while loop
print("Looping pada kolom:")
col_index = 0
columns = df.columns.tolist()
while col_index < len(columns):
    col = columns[col_index]
    print(f"Nama Kolom: {col}")
    print(f"Tipe Data: {df[col].dtype}")
    print(f"Beberapa Nilai: {df[col].head(5).values}")
    print()
    col_index += 1
print('==============================================')

# Looping pada baris untuk 5 data pertama menggunakan while loop
print("Looping pada baris (5 data pertama):")
row_index = 0
max_rows = min(5, len(df))  # Ensure we don't exceed the number of rows in the DataFrame

while row_index < max_rows:
    row = df.iloc[row_index]  # Access row by index
    print(f"Index: {row_index}")
    row_array = row.values  # Convert the row to an array
    print("Data Baris (sebagai array):")
    print(row_array)
    print()
    row_index += 1
```

```python
import pandas as pd

# Membaca dataset dari file CSV
df = pd.read_csv('dataset_day6.csv')

# Menampilkan beberapa baris pertama dari DataFrame untuk eksplorasi
print("Beberapa baris pertama dari DataFrame:")
print(df.head(5))
print()
print('==============================================')

# Looping pada kolom menggunakan nested loop
print("Looping pada kolom dan baris menggunakan nested loop:")
columns = df.columns.tolist()

for col in columns:
    print(f"Nama Kolom: {col}")
    print(f"Tipe Data: {df[col].dtype}")
    print(f"Beberapa Nilai dari Kolom {col}:")

    # Nested loop untuk iterasi baris-baris pada kolom saat ini
    for index in range(min(5, len(df))):  # Looping untuk 5 baris pertama atau kurang jika df lebih kecil dari 5
        print(df[col].iloc[index])

    print()
print('==============================================')

# Looping pada baris menggunakan nested loop
print("Looping pada baris (5 data pertama) dan kolom:")
for index in range(min(5, len(df))):  # Looping untuk 5 baris pertama atau kurang jika df lebih kecil dari 5
    row = df.iloc[index]
    print(f"Index: {index}")
    print("Data Baris (sebagai list):")

    # Nested loop untuk iterasi kolom-kolom pada baris saat ini
    row_list = row.tolist()
    for value in row_list:
        print(value)

    print()
```
