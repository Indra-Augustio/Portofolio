### 1. Soal : Tiap Perusahaan ingin menghitung suatu pesanan pada tiap harinya yang memiliki status rejected.
#### (poin 15)
Saran pengerjaan :

Anda memiliki tabel orders dengan kolom-kolom berikut: company, order_date, status, dan order_id.

Tulislah sebuah query SQL yang menghasilkan:

1. Kolom company
2. Kolom tanggal yang merupakan hari dari tanggal order (dalam format tanggal dua digit).
3. Kolom total_rejected yang merupakan jumlah order yang memiliki status 'REJECTED' pada setiap perusahaan (company) pada setiap tanggal order.

Output dari query tersebut harus mengelompokkan hasil berdasarkan company dan tanggal order (tanggal), serta menghitung jumlah order yang memiliki status 'REJECTED' untuk setiap kelompok.

```python
import pandas as pd
import sqlite3

# Memuat dataset ke dalam DataFrame
orders = pd.read_excel('orders.xlsx')

# Ganti '/path/to/' dengan path yang benar ke file 'order_detail.xlsx'
order_details = pd.read_excel('order_detail.xlsx')

# Membuat koneksi ke database SQLite
conn = sqlite3.connect(':memory:')

# Menulis DataFrame ke dalam tabel database
orders.to_sql('orders', conn, if_exists='replace', index=False)
order_details.to_sql('order_details', conn, if_exists='replace', index=False)

# Menjalankan query SQL
query = """
SELECT
    o.company,
    strftime('%d', o.order_date) AS order_day,
    COUNT(CASE WHEN o.status = 'REJECTED' THEN o.order_id ELSE NULL END) AS total_rejected
FROM
    orders o
GROUP BY
    o.company, order_day
ORDER BY
    o.company, order_day;
"""

# Eksekusi query dan tampilkan hasilnya
df_result = pd.read_sql_query(query, conn)
print(df_result)

# Menutup koneksi
conn.close()
```

### 2. Atasan anda ingin mengetahui apa saja produk yang telah dipesan oleh setiap pelanggan (customer_id) bersama dengan jumlahnya (quantity) pada setiap pesanan?
#### (poin 15)
Saran pengerjaan :
Anda memiliki dua tabel, yaitu orders dan order_details, dengan kolom-kolom berikut:

 1.   Tabel orders memiliki kolom customer_id, order_id, dan mungkin kolom lain terkait pesanan.
 2.   Tabel order_details memiliki kolom order_id, product_name, quantity, dan mungkin kolom lain terkait detail produk.

Tuliskan sebuah query SQL yang melakukan hal berikut:

 1.   Memilih kolom customer_id, order_id, product_name, dan quantity dari kedua tabel tersebut.
 2.   Menggabungkan kedua tabel berdasarkan order_id.
 3.   Mengurutkan hasil berdasarkan customer_id dan order_id.

```python
import pandas as pd
import sqlite3

# Membuat koneksi ke database SQLite
conn = sqlite3.connect(':memory:')

# Menulis DataFrame ke dalam tabel database
orders.to_sql('orders', conn, if_exists='replace', index=False)
order_details.to_sql('order_details', conn, if_exists='replace', index=False)

# Menjalankan query SQL
query = """
SELECT
    o.customer_id,
    o.order_id,
    od.product_name,
    od.quantity
FROM
    orders o
JOIN
    order_details od ON o.order_id = od.order_id
ORDER BY
    o.customer_id, o.order_id;
"""

# Eksekusi query dan tampilkan hasilnya
df_result = pd.read_sql_query(query, conn)
print(df_result)

# Menutup koneksi
conn.close()
```

### 3. Seperti biasa, atasa anda ingin menampilkan total orders dan kategori di tiap perusahaan dengan beberapa kriteria berikut :
Saran Pengerjaan :
1. category harus :
2. Good (total order status sent > 2 dan total rejected > 0),
3. Superb (tidak ada yang di reject, dan total order status sent > 2),
4. selain ini dia Normal

### notes :
1. Kerjakan secara CTE **(poin 15)**
2. kerjakan secara Nested Query **(poin 15)**
3. Harusnya hasilnya sama, Apakah hasilnya sama setelah anda running jawaban pada poin 1 dan 2 diatas ?

```python
import pandas as pd
import sqlite3

# Memuat dataset ke dalam DataFrame
orders = pd.read_excel('orders.xlsx')

# Membuat koneksi ke database SQLite
conn = sqlite3.connect(':memory:')

# Menulis DataFrame ke dalam tabel database
orders.to_sql('orders', conn, if_exists='replace', index=False)

# 1. Menggunakan CTE (Common Table Expression)
print("Hasil dengan CTE:")
query_cte = """
WITH CategoryCounts AS (
  SELECT
    company,
    SUM(CASE WHEN status = 'SENT' THEN 1 ELSE 0 END) AS total_sent,
    SUM(CASE WHEN status = 'REJECTED' THEN 1 ELSE 0 END) AS total_rejected
  FROM orders
  GROUP BY company
)
SELECT
  cc.company,
  cc.total_sent,
  cc.total_rejected,
  CASE
    WHEN cc.total_sent > 2 AND cc.total_rejected > 0 THEN 'Good'
    WHEN cc.total_sent > 2 AND cc.total_rejected = 0 THEN 'Superb'
    ELSE 'Normal'
  END AS category
FROM CategoryCounts cc;
"""
df_result_cte = pd.read_sql_query(query_cte, conn)
print(df_result_cte)

# 2. Menggunakan Nested Query
print("\nHasil dengan Nested Query:")
query_nested = """
SELECT
  o.company,
  (SELECT COUNT(*) FROM orders WHERE company = o.company AND status = 'SENT') AS total_sent,
  (SELECT COUNT(*) FROM orders WHERE company = o.company AND status = 'REJECTED') AS total_rejected,
  CASE
    WHEN (SELECT COUNT(*) FROM orders WHERE company = o.company AND status = 'SENT') > 2 AND (SELECT COUNT(*) FROM orders WHERE company = o.company AND status = 'REJECTED') > 0 THEN 'Good'
    WHEN (SELECT COUNT(*) FROM orders WHERE company = o.company AND status = 'SENT') > 2 AND (SELECT COUNT(*) FROM orders WHERE company = o.company AND status = 'REJECTED') = 0 THEN 'Superb'
    ELSE 'Normal'
  END AS category
FROM orders o
GROUP BY o.company;
"""
df_result_nested = pd.read_sql_query(query_nested, conn)
print(df_result_nested)

# Menutup koneksi
conn.close()
```
